```python
# import seaborn as sns
import pandas as pd
```

```python
# 读取数据
data1 = pd.read_excel("C:/Users/20851/Desktop/2022年MathorCup大数据竞赛-赛道B初赛/附件1语音业务用户满意度数据.xlsx")

print(list(data1.columns))
```

```python
#数据清洗
#删除5列无用数据
del data1['用户id']
del data1['用户描述']
del data1['用户描述.1']
del data1['终端品牌']
del data1['终端品牌类型']

#对数据进行统一编码，前3列不用动
data1.iloc[:,4] = data1.iloc[:,4].map({1:1,2:0})
for i in range(5,20):
    data1.iloc[:,i] = data1.iloc[:,i].map({-1:0,1:1,2:1,3:1,4:1,5:1,6:1,7:1,98:1})

data1.iloc[:,23] = data1.iloc[:,23].fillna(0)
data1.iloc[:,24] = data1.iloc[:,24].fillna(0)
data1.iloc[:,27] = data1.iloc[:,27].map({'2G':2,'4G':4,'5G':5})
data1.iloc[:,28] = data1.iloc[:,28].map({'CSFB':0, 'EPSFB':1, 'GSM':2,'VOLTE':3,'VONR':4})
for i in ['是否关怀用户', '是否去过营业厅','是否4G网络客户（本地剔除物联网）', '是否5G网络客户', '是否实名登记用户']:
    data1[i] = data1[i].fillna(0).map({0:0,'是':1,'否':0})#29,30,33,45,46
data1.iloc[:,42] = data1.iloc[:,42].fillna(0)
data1.iloc[:,47] = data1.iloc[:,47].fillna(0).map({'未评级':0,'准星':0, '一星':1,'二星':2,'三星':3, '白金卡':4, '钻石卡':5, '银卡':6,'金卡':7})

data1 = data1.dropna()#删除后面的缺失行
print(data1.head(5))

#保存成excel文件
# data1.to_excel("C:/Users/20851/Desktop/2022年MathorCup大数据竞赛-赛道B初赛/清洗后语音满意度数据.xlsx",index=None)
```

```python
del data1['网络覆盖与信号强度']
del data1['语音通话清晰度']
del data1['语音通话稳定性']

#获取相关系数矩阵
par = data1.corr().iloc[:,[0]].apply(abs).sort_values('语音通话整体满意度',ascending=0)
print(par)

#各属性相关性的热力图
# sns.heatmap(data1.corr(),linewidths=0.1,vmax=1.0, square=True,linecolor='white', annot=True)
# ['是否遇到过网络问题','居民小区','手机没有信号','有信号无法拨通','通话过程中突然中断','办公室','通话过程中一方听不见','通话中有杂音、听不清、断断续续']
```
